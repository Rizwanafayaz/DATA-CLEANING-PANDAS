```python
import pandas as pd
df = pd.read_excel('Customer Call list.xlsx')
```

```python
df.head()
```

```python
#removing duplicates
df = df.drop_duplicates()
df
```

```python
# Standardizing Last_Name Column
df["Last_Name"] = df["Last_Name"].str.lstrip("...")
df["Last_Name"] = df["Last_Name"].str.lstrip("/")
df["Last_Name"] = df["Last_Name"].str.rstrip("_")
df["Last_Name"] = df["Last_Name"].str.strip("123._/")
```

```python
df.head()
```

```python
#getting rid of unncessary columns
df = df.drop(columns = "Not_Useful_Column")
df.head()
```

```python
# Standardizing Phone_Number Column
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]', '', regex=True)
```

```python
df.head()
```

```python
# Converting Data Type of Phone_Number Column to string
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
```

```python
#### Formatting Phone_Number Column
# Applied a lambda function to format `Phone_Number` column into a specific pattern with hyphens (`-`) after every three characters.

df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
df.head(28)
```

```python
# Handling NaN/NA Values in Phone_Number column
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','')

df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','')
```

```python
df.head(24)
```

```python
# Splitting Address into Street_Address, State, and Zip_Code Columns
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',', n=2, expand=True)
```

```python
df.head()
```

```python
# Mapping 'Yes' and 'No' to 'Y' and 'N' in Do_Not_Contact Column
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')

df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')
```

```python
df.head()
```

```python
# Mapping 'Yes' and 'No' to 'Y' and 'N' in Paying Customer Column
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes','Y')

df["Paying Customer"] = df["Paying Customer"].str.replace('No','N')
```

```python
df.head()
```

```python
# Filling Missing Values with Empty String
df=df.fillna('')
```

```python
df.head()
```

```python
# Removing Rows where Do_Not_Contact is 'Y'
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)
```

```python
df.head()
```

```python
# Removing Rows with Empty Phone Numbers
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace=True)
```

```python
df.head()
```

```python
# Resetting DataFrame Index
df = df.reset_index(drop=True)
```
